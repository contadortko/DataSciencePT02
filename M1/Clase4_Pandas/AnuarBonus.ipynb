"""A partir del CSV hospitales2.csv, generar un archivo CSV de salida, que contenga
los siguientes campos en este orden:
latitude
longitude
name
label
Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder 
con la dirección del hospital, y el campo label con el nombre del hospital."""

```python
# Importar la biblioteca pandas
import pandas as pd
```

```python


# Leer el archivo CSV "hospitales2.csv" y crear un DataFrame llamado "df_Hosp", además de separar por , cada serie o columna
df_Hosp = pd.read_csv('hospitales2.csv', sep=",")

# Eliminar la cadena "POINT" de la columna "WKT" en cada fila del DataFrame
df_Hosp["WKT"] = df_Hosp.WKT.str.replace("POINT", "")

# Eliminar los paréntesis de apertura y cierre en la columna "WKT" en cada fila del DataFrame
df_Hosp["WKT"] = df_Hosp.WKT.str.replace(")", "")
df_Hosp["WKT"] = df_Hosp.WKT.str.replace("(", "")

# Renombrar la columna "WKT" a "Coordenadas"
df_Hosp.rename(columns = {"WKT": "Coordenadas"}, inplace = True)


```

```python
# Mostrar las dos primeras filas del DataFrame resultante
df_Hosp.head(2)
```

```python
df_Hosp[[' ','Latitud', 'Longitud']] = df_Hosp['Coordenadas'].str.split(' ', expand=True)
del df_Hosp[" "]
del df_Hosp["Coordenadas"]
```

```python
df_Hosp.rename(columns = {"DOM_NORMA": "Name", "NOMBRE":"Label"}, inplace = True)
df_Hosp.head()
```

```python
df_HospNew = df_Hosp.reindex(columns = ['Latitud', 'Longitud','Name', 'Label','NOM_MAP', 'ID', 'OBJETO', 'CALLE', 'ALTURA',  'DOM_GEO', 'TELEFONO', 'GUARDIA', 'FAX', 'WEB', 'COD_POSTAL', 'TIPO', 'TIPO_ESPEC', 'MOD_AT_1', 'MOD_AT_2', 'DEPEND_ADM', 'DIRECTOR'])
df_HospNew.head()
```

```python
df_HospNew = df_HospNew.iloc[:,0:4]
```
