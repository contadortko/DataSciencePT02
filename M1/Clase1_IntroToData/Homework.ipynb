2) Convertir de decimal a binario las fracciones 1/2, 1/3, 1/4, 1/5, 1/6, 1/7, 1/8, 1/9. Luego analizar los resultados y observar qué particularidad se encuentra en los mismos. Se puede usar Python o una calculadora, lo importante es ver si hay algo que podemos notar...

```python
from fractions import Fraction #Importa la clase Fraction de la biblioteca fractions

def DecimalBinario(numeroFraccion): #Define la función que convierte un número fraccionario a su representación Binaria
     
   entero = int(numeroFraccion) #Extrae la parte entera del numero fraccionario
   decimal = (numeroFraccion - entero) #extrae la parte decimal del numero fraccionario
   #Inicializa los resultados binarios de la parte entera y decimal
   resultado_entero = "" 
   resultado_decimal = ""

#convertir la parte entera a binario
   while entero > 0 :
      resultado_entero += str(entero%2) #agrega el residuo de la división entre 2 al resultado binario
      entero //= 2 #divide la parte entera entre dos devolviendo su división entera
   if entero == 0:
      resultado_entero = "0"
#convertir la parte fraccionaria binario
   while decimal > 0 :
      decimal *= 2 #Multiplica la parte decimal por 2
      if decimal >= 1: #Si la parte decimal resultante es mayor o igual a 1, agrega 1 al resultado binario
         resultado_decimal += "1"
         decimal -= 1 #resta un 1 de la parte decimal si se agrega  un 1  al resultado binario.
      else:
         resultado_decimal += "0" #Sino es mayor a 1 agrega un 0 al resultado binario
   
#presentación binaria truncada a 24 digitos
   return f"Fracción {Fraction(numeroFraccion).limit_denominator()}: {round(numeroFraccion,5)} y en Binario: {resultado_entero}.{resultado_decimal[:24]}"

resultado = DecimalBinario(1/9)
print(resultado) 
            
```
